Numeric play with a problem presented with an analytic solution.

*   The original question, which is larger 50 to the power 99 or 50 factorial.
*   The analytic way is interesting.

*   This bit of code shows that power is always larger using v from 0..99
*   But this changes using v from 0..150 at v=133


*   Also, leading on from this, algorithms used to multiply larger numbers (see notes an excellent summary "The Genius Way Computers Multiply Big Numbers" by PurpleMind. O(n log n) is best by Harvey-Hoeven (2025)

*   Python uses a mixture of older algorithms







```python
import math


for v in range(150):
    print(f'v {v}')
    # factorial
    f99 = math.factorial(v)
    print(f'f99 {f99}')
    # power
    p99 = 50**v
    print(f'p99 {p99}')
    if f99 > p99:
        print('factorial larger')
        #break
    else:
        print('power larger')

    print('ratio:',f99/p99)

```
